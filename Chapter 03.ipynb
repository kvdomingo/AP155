#### Exercise 3.2 Curve plotting

```python
theta = np.linspace(0,2*np.pi)
x = 2*np.cos(theta) + np.cos(2*theta)
y = 2*np.sin(theta) - np.sin(2*theta)
mp.plot(x,y, "r:", label="Deltoid")
mp.title("Deltoid")
mp.xlabel("$x$")
mp.ylabel("$f(x)$")
mp.show()

theta = np.linspace(0,10*np.pi,1000)
r = theta**2
x = r*np.cos(theta)
y = r*np.sin(theta)
mp.plot(x,y, "c-.", label="Galilean spiral")
mp.title("Galilean spiral")
mp.xlabel("$x$")
mp.ylabel("$f(x)$")
mp.show()

theta = np.linspace(0,24*np.pi,100000)
r = np.exp(np.cos(theta)) - 2*np.cos(4*theta) + (np.sin(theta/12))**5
x = r*np.cos(theta)
y = r*np.sin(theta)
mp.plot(x,y, "m", label="Fey's function")
mp.title("Fey's function")
mp.xlabel("$x$")
mp.ylabel("$f(x)$")
mp.show()
```

#### Example 3.1 Wave Interference

```python
wavelength = 5.0
k = 2*np.pi/wavelength
xi0 = 1.0
sep = 20.0
side = 100.0
points = 500
spacing = side/points

x1 = side/2 + sep/2
y1 = side/2
x2 = side/2 - sep/2
y2 = side/2

xi = np.empty([points,points],float)

for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r1 = np.sqrt((x-x1)**2+(y-y1)**2)
        r2 = np.sqrt((x-x2)**2+(y-y2)**2)
        xi[i,j] = xi0*np.sin(k*r1) + xi0*np.sin(k*r2)

mp.imshow(xi,origin="lower",extent=[0,side,0,side])
mp.gray()
mp.show()
```
